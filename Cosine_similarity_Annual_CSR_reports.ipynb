```
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from scipy import sparse
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
def getcsmatrix(l):
  temp=[list(i) for i in l]
  A =  np.array(temp)
  A_sparse = sparse.csr_matrix(A)
  similarities_sparse = cosine_similarity(A_sparse,dense_output=False)
  return similarities_sparse.toarray()
```

```
def get_man_df(df_mandated):
  l=[]
  df_mandated=df_new[df_new['dummy_criteria_csr_avg']==1]
  vectorizer=CountVectorizer()
  y=vectorizer.fit(df_mandated['Text'])
  for i in df_mandated['Text']:
    r=vectorizer.transform([i])
    l.append(r)
  x=[]
  for i in l:
    x.append(i.toarray())
  df_mandated['Count_vectors']=x
  l=[]
  for i in range(df_mandated.shape[0]):
    l.append(df_mandated['Count_vectors'].iloc[i][0])
  df_mandated['Final_count_vectors']=l
  df_mandated
  man_matrix=getcsmatrix(df_mandated['Final_count_vectors'])
  l=[]
  for i in range(len(man_matrix)):
    l.append(man_matrix[i])
  df_mandated['similarity']=l
  means=[]
  for i in df_mandated['similarity']:
    mean=(i.sum()-1)/(len(i)-1)
    means.append(mean)
  df_mandated['means']=means
  return df_mandated
```
