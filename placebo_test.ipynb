```python
rural=[]
semi_urban=[]
urban=[]
others_len=[]
sfbs=[]
banks=[]
nbfc_mfi=[]
for i in range(mfi.shape[0]):
    if(mfi.ru_classification.iat[i]=='Rural'):
        rural.append(1)
        semi_urban.append(0)
        urban.append(0)
    elif(mfi.ru_classification.iat[i]=='Urban'):
        rural.append(0)
        semi_urban.append(0)
        urban.append(1)
    elif(mfi.ru_classification.iat[i]=='Semi-Urban'):
        rural.append(0)
        semi_urban.append(1)
        urban.append(0)
    else:
        rural.append(0)
        semi_urban.append(0)
        urban.append(0)
len_type=['Others', 'SFBs', 'Banks', 'NBFC MFI']
for i in range(mfi.shape[0]):
    if(mfi.lendertype.iat[i]==len_type[0]):
        others_len.append(1)
        sfbs.append(0)
        nbfc_mfi.append(0)
        banks.append(0)
    elif(mfi.lendertype.iat[i]==len_type[1]):
        others_len.append(0)
        sfbs.append(1)
        nbfc_mfi.append(0)
        banks.append(0)
    elif(mfi.lendertype.iat[i]==len_type[2]):
        others_len.append(0)
        sfbs.append(0)
        nbfc_mfi.append(0)
        banks.append(1)
    elif(mfi.lendertype.iat[i]==len_type[3]):
        others_len.append(0)
        sfbs.append(0)
        nbfc_mfi.append(1)
        banks.append(0)
    else:
        others_len.append(0)
        sfbs.append(0)
        nbfc_mfi.append(0)
        banks.append(0)
```

```python
def get_dates(df):
    l=[]
    for i in range(df.shape[0]):
        
        year=df['qrt_yr'].iat[i][0:4]
        if(df['qrt_yr'].iat[i][4:]=='Q1'):
            l.append(year+'-01-01')
        elif(df['qrt_yr'].iat[i][4:]=='Q2'):
            l.append(year+'-04-01')
        elif(df['qrt_yr'].iat[i][4:]=='Q3'):
            l.append(year+'-07-01')
        elif(df['qrt_yr'].iat[i][4:]=='Q4'):
            l.append(year+'-10-01')
        else:
            l.append(None)
    df['date']=pd.to_datetime(l)    
    return df
```

```python
plac_states=['CG','UP','HR','JH','BR','AS','SK']
```

```python
import random
def get_rand_3_states(l):
    return random.sample(l,3)

def get_bord_districts(df,test):
    return df[(df[test[0]]==1)|(df[test[1]]==1)|(df[test[2]]==1)]

def get_non_imp(df):
    return df[(df['state']!='OR')&(df['state']!='WB')&(df['state']!='DL')]

def get_fk_df(df,test):
    t=[]
    for i in range(df.shape[0]):
        if(df['state'].iat[i]==test[0] or df['state'].iat[i]==test[1] or df['state'].iat[i]==test[2]):
            t.append(0)
        else:
            t.append(df.int_ay.iat[i])
    df['int_ay']=t
    return df

from linearmodels import PanelOLS
def vol_reg(df):
    df=df.set_index(['district','date'])
    df=df[df['ru_classification']!='']
    df=df[df['lendertype']!='']
    df['const']=1
    mod = PanelOLS(df['volume_default_90_plus'],df[['int_ay','urban', 'dummy_semi_urban', 'dummy_sfbs', 'dummy_banks', 'dummy_nbfcs','const']],entity_effects=True,time_effects=True)
    res = mod.fit(cluster_entity=True,cov_type='clustered')
    return [res.params[0],res.pvalues[0],res.tstats[0]]
```

```python

mainl=[]
c=0
for i in range(100):
    c+=1
    non_imp=get_non_imp(mfi_bord)
    test=get_rand_3_states(plac_states)
    temp=get_bord_districts(non_imp,test)
    reg_df=get_fk_df(temp,test)
    coefs_p_t=vol_reg(reg_df)
    coefs_p_t.append(test)
    mainl.append(coefs_p_t)
    print(coefs_p_t,c)
    
```

```python
reg_res1=pd.DataFrame(mainl,columns=['coef','p','t','state_comb']).drop_duplicates('coef')
reg_res1.shape[0]
```
