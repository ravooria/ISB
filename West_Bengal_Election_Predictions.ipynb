```
#1 represents other max_vote,2 represents inc_left
prob_inc=[]
prob_others=[]
prob_aitmc=[]
for i in range(data.shape[0]):
  if(data['inc_others_dummy_bjp_pred_0_new'].iloc[i]==0):
    prob_inc.append(0)
    prob_others.append(0)
    prob_aitmc.append(data['prob_0'].iloc[i])
  if(data['inc_others_dummy_bjp_pred_0_new'].iloc[i]==1):
    prob_others.append(data['prob_0'].iloc[i])
    prob_inc.append(0)
    prob_aitmc.append(0)
  if(data['inc_others_dummy_bjp_pred_0_new'].iloc[i]==2):
    prob_inc.append(data['prob_0'].iloc[i])
    prob_others.append(0)
    prob_aitmc.append(0)
data['prob_inc']=prob_inc
data['prob_aitmc']=prob_aitmc
data['prob_others']=prob_others
data
```

```
def get_value(df):
  df['pred_vote_share_new_bjp']=df['prob_1']*df['total_votes']
  df['pred_vote_share_new_aitmc']=(df['prob_aitmc'])*df['total_votes']
  df['pred_vote_share_new_inc']=(df['prob_inc'])*df['total_votes']
  df['pred_vote_share_new_others']=(df['prob_others'])*df['total_votes']
  df['chance_bjp']=df['pred_vote_share_new_bjp'].sum()/df['total_votes'].sum()
  df['chance_aitmc']=df['pred_vote_share_new_aitmc'].sum()/df['total_votes'].sum()
  df['chance_inc']=df['pred_vote_share_new_inc'].sum()/df['total_votes'].sum()
  df['chance_others']=df['pred_vote_share_new_others'].sum()/df['total_votes'].sum()
  return df
```
